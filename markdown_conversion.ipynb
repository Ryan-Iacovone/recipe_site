```python
from pathlib import Path
from datetime import date

# --------------------------------------------------
# CONFIG
# --------------------------------------------------

RECIPE_DIR = Path("Recipes")  # change this

YAML_FRONT_MATTER = f"""---
category: Dinner
cuisine: Italian
cook_time_minutes:
tags: []
last_modified: {date.today()}
---
"""

# --------------------------------------------------
# PROCESS FILES
# --------------------------------------------------

def add_yaml_front_matter(md_file: Path):
    content = md_file.read_text(encoding="utf-8")

    # Skip files that already contain front matter
    if content.lstrip().startswith("---"):
        print(f"Skipping (already has YAML): {md_file.name}")
        return

    new_content = YAML_FRONT_MATTER + content
    md_file.write_text(new_content, encoding="utf-8")

    print(f"Updated: {md_file.name}")


def main():
    md_files = RECIPE_DIR.glob("*.md")

    for md_file in md_files:
        add_yaml_front_matter(md_file)


if __name__ == "__main__":
    main()
```
